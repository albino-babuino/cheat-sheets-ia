# Pandas - DataFrames y Series

Este notebook cubre los conceptos fundamentales de Pandas: creación de Series y DataFrames, y sus propiedades básicas.

## Importar Pandas


```python
import pandas as pd
import numpy as np
```

## Crear Series


```python
# Desde lista
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])

# Desde diccionario
s3 = pd.Series({'idx1': 1, 'idx2': 2, 'idx3': 3})  # {'idx1': val1, 'idx2': val2}

# Desde NumPy array
s4 = pd.Series(np.array([1, 2, 3, 4]))

# Con nombre
s5 = pd.Series([1, 2, 3, 4], name='mi_serie')

# Crear Series con valores, índice y nombre
s6 = pd.Series([1, 2, 3], index=['a', 'b', 'c'], name='mi_serie')

# Propiedades
s1.values      # Array NumPy con valores
s1.index       # Índice (array-like row labels)
s1.name        # Nombre de la serie
s1.dtype       # Tipo de datos
len(s1)        # Longitud (número de filas)
s1.unique()    # Series de valores únicos
```

## Crear DataFrames


```python
# Desde diccionario (cada columna es una Series)
df1 = pd.DataFrame({
    'nombre': ['Ana', 'Juan', 'María'],
    'edad': [25, 30, 28],
    'ciudad': ['Madrid', 'Barcelona', 'Valencia']
})

# Desde diccionario de Series (comparten el mismo índice)
df1_alt = pd.DataFrame({
    'col1': pd.Series([1, 2, 3], index=['a', 'b', 'c']),
    'col2': pd.Series([4, 5, 6], index=['a', 'b', 'c'])
})

# Desde lista de listas (2D array)
df2 = pd.DataFrame([[1, 2, 3], [4, 5, 6]], 
                   columns=['A', 'B', 'C'],
                   index=['a', 'b'])

# Desde lista de diccionarios
df3 = pd.DataFrame([
    {'nombre': 'Ana', 'edad': 25},
    {'nombre': 'Juan', 'edad': 30}
])

# Desde NumPy array
df4 = pd.DataFrame(np.random.randn(3, 4), 
                   columns=['A', 'B', 'C', 'D'],
                   index=['a', 'b', 'c'])

# Propiedades básicas
df1.shape      # (n_rows, m_cols) - tupla con dimensiones
df1.size       # Número total de elementos
df1.columns    # Array-like column labels
df1.index      # Array-like row labels
df1.values     # Numpy array con los datos
df1.dtypes     # Tipo de cada columna (Series)
# df1.dtype      # No existe en DataFrame, usar dtypes (comentado porque causa error)
df1.info()     # Información del DataFrame (memoria, tipos)
df1.head()     # Primeras filas (por defecto 5)
df1.tail()     # Últimas filas (por defecto 5)
df1.describe() # Estadísticas descriptivas
```

## Verificación de Tipos de Datos

A menudo necesitas verificar qué tipo de objeto tienes antes de realizar operaciones específicas. Aquí están los métodos para comprobar si un objeto es un DataFrame/Series de Pandas, un array de NumPy o una lista de Python.

```python
# Crear ejemplos de diferentes tipos
df_ejemplo = pd.DataFrame({'A': [1, 2, 3]})
serie_ejemplo = pd.Series([1, 2, 3])
array_ejemplo = np.array([1, 2, 3])
lista_ejemplo = [1, 2, 3]

# Verificar si es DataFrame de Pandas
print("¿Es DataFrame?")
print(f"df_ejemplo: {isinstance(df_ejemplo, pd.DataFrame)}")  # True
print(f"serie_ejemplo: {isinstance(serie_ejemplo, pd.DataFrame)}")  # False
print(f"array_ejemplo: {isinstance(array_ejemplo, pd.DataFrame)}")  # False
print(f"lista_ejemplo: {isinstance(lista_ejemplo, pd.DataFrame)}")  # False
print()

# Verificar si es Series de Pandas
print("¿Es Series?")
print(f"df_ejemplo: {isinstance(df_ejemplo, pd.Series)}")  # False
print(f"serie_ejemplo: {isinstance(serie_ejemplo, pd.Series)}")  # True
print(f"array_ejemplo: {isinstance(array_ejemplo, pd.Series)}")  # False
print(f"lista_ejemplo: {isinstance(lista_ejemplo, pd.Series)}")  # False
print()

# Verificar si es array de NumPy
print("¿Es array de NumPy?")
print(f"df_ejemplo: {isinstance(df_ejemplo, np.ndarray)}")  # False
print(f"serie_ejemplo: {isinstance(serie_ejemplo, np.ndarray)}")  # False
print(f"array_ejemplo: {isinstance(array_ejemplo, np.ndarray)}")  # True
print(f"lista_ejemplo: {isinstance(lista_ejemplo, np.ndarray)}")  # False
print()

# Verificar si es lista de Python
print("¿Es lista de Python?")
print(f"df_ejemplo: {isinstance(df_ejemplo, list)}")  # False
print(f"serie_ejemplo: {isinstance(serie_ejemplo, list)}")  # False
print(f"array_ejemplo: {isinstance(array_ejemplo, list)}")  # False
print(f"lista_ejemplo: {isinstance(lista_ejemplo, list)}")  # True
print()

# Verificar si es DataFrame O Series (cualquier objeto de Pandas)
print("¿Es DataFrame o Series?")
print(f"df_ejemplo: {isinstance(df_ejemplo, (pd.DataFrame, pd.Series))}")  # True
print(f"serie_ejemplo: {isinstance(serie_ejemplo, (pd.DataFrame, pd.Series))}")  # True
print(f"array_ejemplo: {isinstance(array_ejemplo, (pd.DataFrame, pd.Series))}")  # False
print()

# Método alternativo: usar type() y comparar
print("Usando type():")
print(f"df_ejemplo: {type(df_ejemplo) == pd.DataFrame}")  # True
print(f"serie_ejemplo: {type(serie_ejemplo) == pd.Series}")  # True
print(f"array_ejemplo: {type(array_ejemplo) == np.ndarray}")  # True
print(f"lista_ejemplo: {type(lista_ejemplo) == list}")  # True
print()

# Función útil para verificar múltiples tipos
def verificar_tipo(objeto):
    """Verifica el tipo de un objeto y devuelve una descripción."""
    if isinstance(objeto, pd.DataFrame):
        return "DataFrame de Pandas"
    elif isinstance(objeto, pd.Series):
        return "Series de Pandas"
    elif isinstance(objeto, np.ndarray):
        return "Array de NumPy"
    elif isinstance(objeto, list):
        return "Lista de Python"
    else:
        return f"Tipo desconocido: {type(objeto)}"

# Ejemplo de uso
print("Función de verificación:")
print(f"df_ejemplo: {verificar_tipo(df_ejemplo)}")
print(f"serie_ejemplo: {verificar_tipo(serie_ejemplo)}")
print(f"array_ejemplo: {verificar_tipo(array_ejemplo)}")
print(f"lista_ejemplo: {verificar_tipo(lista_ejemplo)}")
```
