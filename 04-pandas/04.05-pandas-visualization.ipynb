# Pandas - Visualización de Datos

Este notebook cubre la visualización de datos directamente con Pandas usando el método `.plot()`. Pandas proporciona una interfaz de alto nivel sobre matplotlib que facilita la creación de gráficos comunes sin necesidad de usar matplotlib directamente.

**Nota:** Aunque pandas usa matplotlib internamente, no necesitas importar matplotlib explícitamente para crear gráficos básicos. Sin embargo, para personalización avanzada o para mostrar/guardar gráficos, puedes necesitar `import matplotlib.pyplot as plt`.

## Importar Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Opcional: para personalización avanzada y plt.show()
```

## Crear Datos de Ejemplo

```python
# Series de ejemplo
s = pd.Series([1, 3, 2, 4, 5, 6, 4, 3], index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])

# DataFrame de ejemplo
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [4, 5, 6, 7, 8],
    'C': [2, 3, 4, 5, 6]
}, index=['Ene', 'Feb', 'Mar', 'Abr', 'May'])

# DataFrame con fechas
fechas = pd.date_range('2024-01-01', periods=12, freq='M')
df_temporal = pd.DataFrame({
    'Ventas': np.random.randint(100, 500, 12),
    'Gastos': np.random.randint(50, 300, 12)
}, index=fechas)

# DataFrame para scatter
df_scatter = pd.DataFrame({
    'X': np.random.randn(100),
    'Y': np.random.randn(100),
    'Categoria': np.random.choice(['A', 'B', 'C'], 100)
})

# DataFrame para histograma
df_hist = pd.DataFrame({
    'Valores': np.random.normal(100, 15, 1000)
})
```

## Gráfico de Línea (Line Plot)

El tipo de gráfico por defecto. Útil para visualizar tendencias a lo largo del tiempo.

```python
# Gráfico de línea en Series
s.plot()
plt.show()  # Mostrar gráfico
```

```python
# Gráfico de línea en DataFrame (múltiples series)
df.plot()
plt.show()
```

```python
# Especificar tipo explícitamente
df.plot(kind='line')
plt.show()
```

```python
# Gráfico de línea con datos temporales
df_temporal.plot(kind='line')
plt.show()
```

## Gráfico de Barras (Bar Plot)

Útil para comparar valores categóricos.

```python
# Barras verticales
df.plot(kind='bar')
plt.show()
```

```python
# Barras horizontales
df.plot(kind='barh')
plt.show()
```

```python
# Gráfico de barras apiladas
df.plot(kind='bar', stacked=True)
plt.show()
```

```python
# Gráfico de barras en Series
s.plot(kind='bar')
plt.show()
```

## Histograma (Histogram)

Útil para visualizar la distribución de datos numéricos.

```python
# Histograma en Series
s.plot(kind='hist')
plt.show()
```

```python
# Histograma en DataFrame (una columna)
df_hist['Valores'].plot(kind='hist', bins=30)
plt.show()
```

```python
# Histograma con múltiples columnas (apilado)
df.plot(kind='hist', alpha=0.7, bins=10)
plt.show()
```

## Gráfico de Densidad (KDE - Kernel Density Estimation)

Muestra la estimación de la densidad de probabilidad de una variable continua.

```python
# Gráfico de densidad en Series
df_hist['Valores'].plot(kind='kde')
plt.show()
```

```python
# Gráfico de densidad en DataFrame
df.plot(kind='kde')
plt.show()
```

## Box Plot (Diagrama de Cajas)

Muestra la distribución de datos mediante cuartiles, útil para identificar outliers.

```python
# Box plot en Series
df_hist['Valores'].plot(kind='box')
plt.show()
```

```python
# Box plot en DataFrame (una caja por columna)
df.plot(kind='box')
plt.show()
```

## Gráfico de Dispersión (Scatter Plot)

Útil para visualizar la relación entre dos variables numéricas.

```python
# Scatter plot (requiere especificar x e y)
df_scatter.plot(kind='scatter', x='X', y='Y')
plt.show()
```

```python
# Scatter plot con tamaño variable (s)
df_scatter['Tamaño'] = np.random.randint(10, 100, 100)
df_scatter.plot(kind='scatter', x='X', y='Y', s=df_scatter['Tamaño'], alpha=0.6)
plt.show()
```

## Gráfico Circular (Pie Chart)

Útil para mostrar proporciones de una categoría.

```python
# Pie chart en Series
s.plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
# Pie chart en DataFrame (especificar columna con y)
df_pie = pd.DataFrame({
    'Valores': [20, 30, 25, 15, 10]
}, index=['A', 'B', 'C', 'D', 'E'])
df_pie.plot(kind='pie', y='Valores', autopct='%1.1f%%')
plt.show()
```

## Gráfico de Área (Area Plot)

Similar al gráfico de línea pero con el área bajo la curva rellena.

```python
# Gráfico de área
df.plot(kind='area')
plt.show()
```

```python
# Gráfico de área apilado
df.plot(kind='area', stacked=True, alpha=0.7)
plt.show()
```

## Hexbin Plot

Útil para visualizar la relación entre dos variables cuando hay muchos puntos de datos.

```python
# Hexbin plot (requiere x e y)
df_scatter.plot(kind='hexbin', x='X', y='Y', gridsize=20)
plt.show()
```

## Personalización de Gráficos

Pandas permite personalizar los gráficos directamente con parámetros del método `.plot()`.

```python
# Título y etiquetas de ejes
df.plot(kind='line', title='Ventas Mensuales', xlabel='Mes', ylabel='Cantidad')
plt.show()
```

```python
# Tamaño de la figura
df.plot(kind='bar', figsize=(10, 6))
plt.show()
```

```python
# Colores personalizados
df.plot(kind='bar', color=['red', 'blue', 'green'])
plt.show()
```

```python
# Transparencia (alpha)
df.plot(kind='area', alpha=0.5)
plt.show()
```

```python
# Estilo de línea y marcadores
df.plot(kind='line', style=['-', '--', '-.'], marker='o', markersize=8)
plt.show()
```

```python
# Rotación de etiquetas del eje x
df.plot(kind='bar', rot=45)
plt.show()
```

```python
# Límites de ejes
df.plot(kind='line', xlim=(0, 4), ylim=(0, 10))
plt.show()
```

```python
# Leyenda personalizada
ax = df.plot(kind='line', legend=True)
ax.legend(loc='upper left', title='Columnas')
plt.show()
```

```python
# Grid (cuadrícula)
df.plot(kind='line', grid=True)
plt.show()
```

## Subplots (Múltiples Gráficos)

Puedes crear múltiples gráficos en una misma figura usando el parámetro `subplots`.

```python
# Subplots verticales (una columna por gráfico)
df.plot(kind='line', subplots=True, figsize=(8, 10))
plt.tight_layout()
plt.show()
```

```python
# Subplots con layout personalizado
df.plot(kind='bar', subplots=True, layout=(1, 3), figsize=(15, 5))
plt.tight_layout()
plt.show()
```

## Seleccionar Columnas para Graficar

Puedes especificar qué columnas graficar usando el parámetro `y` o seleccionando columnas antes de graficar.

```python
# Graficar solo columnas específicas
df[['A', 'B']].plot(kind='line')
plt.show()
```

```python
# Graficar una sola columna
df['A'].plot(kind='bar')
plt.show()
```

## Guardar Gráficos

Puedes guardar los gráficos directamente usando el método `.savefig()` del objeto Axes que retorna `.plot()`.

```python
# Guardar gráfico
ax = df.plot(kind='bar', title='Mi Gráfico')
ax.figure.savefig('grafico.png', dpi=300, bbox_inches='tight')
print("Gráfico guardado como 'grafico.png'")
```

```python
# Guardar en diferentes formatos
ax = df.plot(kind='line')
ax.figure.savefig('grafico.pdf')  # PDF
ax.figure.savefig('grafico.svg')  # SVG
ax.figure.savefig('grafico.jpg', dpi=150)  # JPG
print("Gráficos guardados en múltiples formatos")
```

## Métodos de Acceso Directo

Pandas también proporciona métodos de acceso directo para algunos tipos de gráficos comunes.

```python
# Métodos directos (equivalente a .plot(kind='...'))
df.plot.line()      # Equivalente a df.plot(kind='line')
df.plot.bar()        # Equivalente a df.plot(kind='bar')
df.plot.barh()       # Equivalente a df.plot(kind='barh')
df.plot.hist()       # Equivalente a df.plot(kind='hist')
df.plot.box()        # Equivalente a df.plot(kind='box')
df.plot.kde()        # Equivalente a df.plot(kind='kde')
df.plot.area()       # Equivalente a df.plot(kind='area')
df.plot.pie(y='A')   # Equivalente a df.plot(kind='pie', y='A')
df.plot.scatter(x='A', y='B')  # Equivalente a df.plot(kind='scatter', x='A', y='B')
df.plot.hexbin(x='A', y='B')   # Equivalente a df.plot(kind='hexbin', x='A', y='B')

# Ejemplo con método directo
df.plot.bar(title='Usando método directo')
plt.show()
```

## Resumen de Tipos de Gráficos

| Tipo | `kind=` | Método Directo | Descripción |
|------|---------|----------------|-------------|
| Línea | `'line'` | `.plot.line()` | Tendencias temporales |
| Barras verticales | `'bar'` | `.plot.bar()` | Comparación categórica |
| Barras horizontales | `'barh'` | `.plot.barh()` | Comparación categórica horizontal |
| Histograma | `'hist'` | `.plot.hist()` | Distribución de datos |
| Box plot | `'box'` | `.plot.box()` | Distribución y outliers |
| KDE/Densidad | `'kde'` | `.plot.kde()` | Estimación de densidad |
| Scatter | `'scatter'` | `.plot.scatter()` | Relación entre dos variables |
| Pie | `'pie'` | `.plot.pie()` | Proporciones |
| Área | `'area'` | `.plot.area()` | Área bajo la curva |
| Hexbin | `'hexbin'` | `.plot.hexbin()` | Densidad de puntos |

## Parámetros Comunes de Personalización

| Parámetro | Descripción | Ejemplo |
|-----------|-------------|---------|
| `title` | Título del gráfico | `title='Mi Gráfico'` |
| `xlabel` | Etiqueta del eje X | `xlabel='Tiempo'` |
| `ylabel` | Etiqueta del eje Y | `ylabel='Valores'` |
| `figsize` | Tamaño de la figura (ancho, alto) | `figsize=(10, 6)` |
| `color` | Color(es) del gráfico | `color=['red', 'blue']` |
| `alpha` | Transparencia (0-1) | `alpha=0.7` |
| `rot` | Rotación de etiquetas del eje X | `rot=45` |
| `grid` | Mostrar cuadrícula | `grid=True` |
| `legend` | Mostrar leyenda | `legend=True` |
| `subplots` | Crear subplots | `subplots=True` |
| `stacked` | Apilar (para bar/area) | `stacked=True` |
| `xlim` | Límites del eje X | `xlim=(0, 10)` |
| `ylim` | Límites del eje Y | `ylim=(0, 100)` |
