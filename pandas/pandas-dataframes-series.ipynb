# Pandas - DataFrames y Series

## Importar Pandas


```python
import pandas as pd
import numpy as np
```

## Crear Series


```python
# Desde lista
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])

# Desde diccionario
s3 = pd.Series({'a': 1, 'b': 2, 'c': 3})

# Desde NumPy array
s4 = pd.Series(np.array([1, 2, 3, 4]))

# Con nombre
s5 = pd.Series([1, 2, 3, 4], name='mi_serie')

# Propiedades
s1.values      # Array NumPy con valores
s1.index       # Índice
s1.name        # Nombre de la serie
s1.dtype       # Tipo de datos
len(s1)        # Longitud
```

## Crear DataFrames


```python
# Desde diccionario
df1 = pd.DataFrame({
    'nombre': ['Ana', 'Juan', 'María'],
    'edad': [25, 30, 28],
    'ciudad': ['Madrid', 'Barcelona', 'Valencia']
})

# Desde lista de listas
df2 = pd.DataFrame([[1, 2, 3], [4, 5, 6]], 
                   columns=['A', 'B', 'C'])

# Desde lista de diccionarios
df3 = pd.DataFrame([
    {'nombre': 'Ana', 'edad': 25},
    {'nombre': 'Juan', 'edad': 30}
])

# Desde NumPy array
df4 = pd.DataFrame(np.random.randn(3, 4), 
                   columns=['A', 'B', 'C', 'D'])

# Propiedades básicas
df1.shape      # (filas, columnas)
df1.size       # Número total de elementos
df1.columns    # Nombres de columnas
df1.index      # Índice
df1.dtypes     # Tipos de datos por columna
df1.info()     # Información del DataFrame
```
