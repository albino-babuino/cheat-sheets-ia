## Buenas Prácticas

1. **Siempre usa entornos virtuales** antes de instalar paquetes
2. **Usa requirements.txt** para documentar dependencias
3. **Especifica versiones** en requirements.txt para reproducibilidad
4. **Actualiza regularmente** pero prueba antes en desarrollo
5. **No uses `sudo pip install`** - usa entornos virtuales en su lugar
6. **Revisa dependencias** con `pip check` periódicamente
