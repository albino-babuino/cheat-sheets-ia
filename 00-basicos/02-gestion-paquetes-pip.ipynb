# Gestión de Paquetes con pipEste notebook contiene una guía completa sobre cómo gestionar paquetes Python usando `pip`, el gestor de paquetes estándar de Python.

## ¿Qué es pip?`pip` (Pip Installs Packages) es el gestor de paquetes estándar de Python. Permite instalar y gestionar paquetes desde PyPI (Python Package Index) y otros repositorios.### Verificar instalación```bashpip --versionpip3 --version  # En sistemas con Python 2 y 3python -m pip --version```

## Instalar Paquetes### Instalar un paquete```bashpip install nombre-paquetepip install numpypip install pandas```### Instalar versión específica```bashpip install numpy==1.24.0pip install "pandas>=2.0.0"pip install "requests<3.0.0"pip install "matplotlib~=3.7.0"  # Compatible con 3.7.x```### Instalar desde requirements.txt```bashpip install -r requirements.txt```### Instalar en modo usuario (sin sudo)```bashpip install --user nombre-paquete```

## Actualizar Paquetes### Actualizar un paquete específico```bashpip install --upgrade nombre-paquetepip install -U nombre-paquete  # Forma corta```### Actualizar pip mismo```bashpip install --upgrade pippython -m pip install --upgrade pip```

## Listar y Buscar Paquetes### Listar paquetes instalados```bashpip list                    # Listar todos los paquetespip list --outdated         # Listar paquetes desactualizadospip list --format=freeze    # Formato requirements.txt```### Mostrar información de un paquete```bashpip show nombre-paquete     # Información detalladapip show -f nombre-paquete  # Incluir lista de archivos```

## Desinstalar Paquetes### Desinstalar un paquete```bashpip uninstall nombre-paquete```### Desinstalar múltiples paquetes```bashpip uninstall paquete1 paquete2 paquete3```### Desinstalar desde requirements.txt```bashpip uninstall -r requirements.txt -y```

## requirements.txt### Generar requirements.txt```bashpip freeze > requirements.txt```### Instalar desde requirements.txt```bashpip install -r requirements.txt```### Ejemplo de requirements.txt```numpy==1.24.0pandas>=2.0.0matplotlib~=3.7.0jupyterrequests<3.0.0```

## Cache y Configuración### Limpiar cache de pip```bashpip cache purge            # Eliminar todo el cachepip cache list             # Listar cache```### Ver configuración de pip```bashpip config list            # Ver configuración actual```

## Buenas Prácticas1. **Siempre usa entornos virtuales** antes de instalar paquetes2. **Usa requirements.txt** para documentar dependencias3. **Especifica versiones** en requirements.txt para reproducibilidad4. **Actualiza regularmente** pero prueba antes en desarrollo5. **No uses `sudo pip install`** - usa entornos virtuales en su lugar6. **Revisa dependencias** con `pip check` periódicamente
