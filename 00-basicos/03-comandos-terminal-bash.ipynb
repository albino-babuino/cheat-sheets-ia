# Comandos de Terminal - BashEste notebook contiene una guía completa de comandos esenciales de terminal/bash para trabajar eficientemente en Linux/Mac.

## Navegación de Directorios```bashpwd                    # Mostrar directorio actual (Print Working Directory)cd /ruta/directorio    # Cambiar a un directorio específicocd ~                   # Ir al directorio homecd ..                  # Subir un nivel (directorio padre)cd -                   # Volver al directorio anteriorcd                     # Ir al directorio home (equivalente a cd ~)```

## Listar Archivos y Directorios```bashls                     # Listar archivos y directoriosls -l                  # Listar con detalles (permisos, tamaño, fecha)ls -a                  # Listar todos los archivos (incluyendo ocultos)ls -la                 # Listar todos con detallesls -lh                 # Listar con detalles y tamaños legibles (human-readable)ls -lt                 # Listar ordenado por fecha (más recientes primero)ls -R                  # Listar recursivamente (incluye subdirectorios)ls directorio/         # Listar contenido de un directorio específico```

## Crear y Eliminar Directorios```bash# Crear directoriosmkdir nombre_directorio              # Crear un directoriomkdir -p ruta/completa/directorio    # Crear directorios anidadosmkdir dir1 dir2 dir3                 # Crear múltiples directorios# Eliminar directoriosrmdir nombre_directorio              # Eliminar directorio vacíorm -r nombre_directorio              # Eliminar directorio y su contenido (recursivo)rm -rf nombre_directorio             # Eliminar forzadamente sin confirmación```⚠️ **Cuidado con `rm -rf`**: Elimina permanentemente sin posibilidad de recuperación.

## Crear y Eliminar Archivos```bash# Crear archivostouch archivo.txt                    # Crear archivo vacío o actualizar timestamptouch archivo1.txt archivo2.txt     # Crear múltiples archivosecho "contenido" > archivo.txt      # Crear archivo con contenido# Eliminar archivosrm archivo.txt                       # Eliminar un archivorm archivo1.txt archivo2.txt         # Eliminar múltiples archivosrm -i archivo.txt                    # Eliminar con confirmación interactivarm *.txt                             # Eliminar todos los archivos .txt```

## Copiar y Mover Archivos```bash# Copiar archivoscp archivo.txt copia.txt             # Copiar archivocp archivo.txt directorio/           # Copiar archivo a un directoriocp -r directorio/ copia_directorio/  # Copiar directorio recursivamentecp archivo*.txt directorio/           # Copiar múltiples archivos con patrón# Mover/renombrar archivosmv archivo.txt nuevo_nombre.txt      # Renombrar archivomv archivo.txt directorio/           # Mover archivo a un directoriomv directorio1/ directorio2/         # Mover/renombrar directorio```

## Ver Contenido de Archivos```bashcat archivo.txt                      # Mostrar todo el contenidoless archivo.txt                     # Ver contenido página por página (q para salir)more archivo.txt                     # Ver contenido página por páginahead archivo.txt                     # Mostrar primeras 10 líneashead -n 20 archivo.txt               # Mostrar primeras 20 líneastail archivo.txt                     # Mostrar últimas 10 líneastail -n 20 archivo.txt               # Mostrar últimas 20 líneastail -f archivo.log                  # Seguir archivo en tiempo real (útil para logs)```

## Buscar Archivos y Contenido```bash# Buscar archivosfind . -name "archivo.txt"           # Buscar archivo en directorio actual y subdirectoriosfind /ruta -name "*.py"              # Buscar todos los archivos .pyfind . -type f -name "*.txt"         # Buscar solo archivos (no directorios)find . -type d -name "directorio"    # Buscar solo directorios# Buscar contenido en archivosgrep "texto" archivo.txt             # Buscar texto en un archivogrep -r "texto" directorio/          # Buscar recursivamente en directoriogrep -i "texto" archivo.txt          # Búsqueda sin distinguir mayúsculas/minúsculasgrep -n "texto" archivo.txt          # Mostrar número de línea```

## Permisos de Archivos```bash# Ver permisosls -l archivo.txt                    # Ver permisos detallados# Cambiar permisoschmod 755 archivo.txt                # Cambiar permisos (rwxr-xr-x)chmod +x script.sh                   # Añadir permiso de ejecuciónchmod -x archivo.txt                 # Quitar permiso de ejecuciónchmod 644 archivo.txt                # Permisos comunes para archivos (rw-r--r--)chmod 755 directorio/                # Permisos comunes para directorios (rwxr-xr-x)```### Explicación de permisos:- **r** (read): lectura (4)- **w** (write): escritura (2)- **x** (execute): ejecución (1)- **755** = usuario: rwx (7), grupo: r-x (5), otros: r-x (5)

## Variables de Entorno```bash# Ver variables de entornoenv                                  # Mostrar todas las variables de entornoecho $PATH                           # Mostrar valor de PATHecho $HOME                           # Mostrar directorio homeecho $USER                           # Mostrar usuario actual# Establecer variables de entornoexport VARIABLE="valor"              # Establecer variable para sesión actualexport PATH=$PATH:/nueva/ruta        # Añadir ruta a PATH```

## Redirección y Pipes```bash# Redirección de salidacomando > archivo.txt                # Redirigir salida a archivo (sobrescribe)comando >> archivo.txt               # Añadir salida al final del archivocomando 2> error.log                 # Redirigir errores a archivocomando > salida.txt 2>&1            # Redirigir salida y errores al mismo archivo# Pipes (tuberías)comando1 | comando2                  # Pasar salida de comando1 a comando2ls -l | grep "archivo"               # Listar y filtrarcat archivo.txt | wc -l              # Contar líneas de archivo```
